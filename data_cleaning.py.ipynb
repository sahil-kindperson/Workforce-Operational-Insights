```python
import pandas as pd
import numpy as np

#1. Load the Data

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition[1].csv')

#2. Data Cleaning and Validation
print (" __initial data check__ ")
print (f"Original shape: {df.shape}")

#Check for duplicates
duplicates = df.duplicated().sum()
print (f"Duplicate rows: {duplicates}")
if duplicates > 0:
  df.drop_duplicates(inplace=True)

#checking for missing values
missing_values = df.isnull().sum().sum()
print (f"Total Missing Values : {missing_values}")

# Drop irrelevant columns (Columns that provide no value for analysis)
columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df.drop (columns = columns_to_drop, inplace=True, errors='ignore')

#3. Data Transformation
# Rename columns for better SQL compatibility (removing spaces if any, though this dataset is usually clean)
df.columns  = [c.replace(' ', '_') for c in df.columns]

# Map 'Attrition' to binary for calculation purposes
df['Attrition_Binary'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# 4. Export Cleaned Data
df.to_csv('Cleaned_HR_Data.csv', index=False)
print ("---Data Cleaning Complete. Export to 'Cleaned_HR_Data' ---")
```
